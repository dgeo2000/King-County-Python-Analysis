```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(12, 12)) # figure size 
plt.matshow(df_dropped.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_dropped.shape[1]), df_dropped.columns, fontsize=10, rotation=45) # x axis labels
plt.yticks(range(df_dropped.shape[1]), df_dropped.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
f.savefig('test.png')
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))
corr_matrix = df_dropped.corr()
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(corr_matrix, annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

The following are the variables that have a strong positive correlation with price.

- grade
- sqft_living
- sqft_above

# Categorical Plot

```python
sns.histplot(df['bathrooms'], bins = 20, kde = True)
```

```python
df.loc[df['bathrooms'] <= 1, 'bathroom_size'] = 'Few'
```

```python
df.loc[(df['bathrooms'] > 1) & (df['bathrooms'] < 3), 'bathroom_size'] = 'Normal'
```

```python
df.loc[df['bathrooms'] >= 3, 'bathroom_size'] = 'Many'
```

```python
df['bathroom_size'].value_counts(dropna = False)
```

```python
sns.set(style="ticks")
g = sns.catplot(x="bathrooms", y="price", hue="bathroom_size", data=df, height = 6, aspect = 1.5)
```

Coloring by the number of bathrooms, the most expensive houses tend to have many bathrooms. The cheapest tend to have few.
