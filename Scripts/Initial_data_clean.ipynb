```python
for col in df.columns:
    # Get the types of each value in the column
    types_in_column = df[col].map(type)
    
    # Check if any type is different from the type of the first value
    mixed_types = types_in_column != types_in_column.iloc[0]
    
    # If there are any mixed types, print the column name
    if mixed_types.any():
        print(f"Column with mixed types: {col}")
```
